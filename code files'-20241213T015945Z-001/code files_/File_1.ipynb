```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("02-15-2018.csv")
df2 = pd.read_csv("02-16-2018.csv")
df3 = pd.read_csv("02-22-2018.csv")
df4 = pd.read_csv("02-28-2018.csv")
df5 = pd.read_csv("03-02-2018.csv")
```

```python
df1['Date'] = "02-15-2018"
df2["Date"] = "02-16-2018"
df3["Date"] = "02-22-2018"
df4["Date"] = "02-28-2018"
df5["Date"] = "03-02-2018"
```

```python
df_combined = pd.concat([df1,df2,df3,df4,df5], ignore_index=True)
```

```python
df_combined['Date'] = pd.to_datetime(df_combined['Date'])
```

```python
# Drop rows where "Flow Byts/s" has null values
df_combined = df_combined.dropna(subset=["Flow Byts/s"])

# Verify that there are no null values left in the "Flow Byts/s" column
print("Missing values after dropping rows:")
print(df_combined["Flow Byts/s"].isnull().sum())
```

```python
object_columns = df_combined.select_dtypes(include=['object']).columns
print("\nColumns with object data type:")
print(object_columns)
```

```python
columns_to_convert = df_combined.columns.difference(['Timestamp', 'Label','Date'])
df_combined[columns_to_convert] = df_combined[columns_to_convert].apply(pd.to_numeric, errors='coerce')

# Check if all specified columns have been converted to numeric types
print("Data types after conversion:")
print(df_combined.dtypes)
```

```python
df_combined = df_combined.drop(columns=["Timestamp"])
```

```python
unique_labels = df_combined["Label"].unique()
print("Unique values in the 'Label' column:")
print(unique_labels)
```

```python
# Map 'Benign' to 0 and all other labels to 1 for anomaly detection
df_combined["Label"] = df_combined["Label"].apply(lambda x: 0 if x == 'Benign' else 1)

# Verify the mapping
label_counts = df_combined["Label"].value_counts()
print("Counts of each label (0 for Benign, 1 for Anomalies):")
print(label_counts)
```

```python
from sklearn.utils import resample

benign_data = df_combined[df_combined['Label'] == 0]
anomaly_data = df_combined[df_combined['Label'] == 1]

# Undersample the majority class
benign_data_resampled = resample(
    benign_data,
    replace=False,  # No replacement to avoid duplicates
    n_samples=2000000,  # Target number of benign samples
    random_state=42  # For reproducibility
)

# Combine the resampled majority class with the minority class
df_combined = pd.concat([benign_data_resampled, anomaly_data])

# Shuffle the combined dataset
df_combined = df_combined.sample(frac=1, random_state=42).reset_index(drop=True)

# Verify the new distribution of labels
print(f"Counts of each label after undersampling:\n{df_combined['Label'].value_counts()}")
```

```python
df_combined.to_csv('df_combined.csv', index=False)
```
